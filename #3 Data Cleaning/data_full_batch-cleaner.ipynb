```python
import pandas as pd
from battle_log_data_preprocessor import main as preprocess_battle_log_main
```

```python
input_path_list = [f'..//#2 Data Storage//scrapped_data//semi_data_trail_part{i}.csv' for i in range(1, 11)]
output_path_list = [f'..//#2 Data Storage//Processed Data//preprocessed_battle_log_part{i}.csv' for i in range(1, 11)]
```

```python
for input_path, output_path in zip(input_path_list, output_path_list):
    preprocess_battle_log_main(input_filename=input_path, output_filename=output_path)
```

```python
# now i want to read all the output file combine them into one dataframe 
df_list = [pd.read_csv(output_path) for output_path in output_path_list]
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
combined_df.head()
```

```python
#drop rows if replay tag is empty

print(f'Number of rows before dropping empty replay tags: {combined_df.shape[0]}')
combined_df.drop(combined_df[combined_df['replayTag'] == ''].index, inplace=True)
#print number of rows after dropping
combined_df.dropna(subset=['replayTag'], inplace=True)
print(f'Number of rows after dropping empty replay tags: {combined_df.shape[0]}')
# drop rows with duplicate replay tags
combined_df.drop_duplicates(subset=['replayTag'], inplace=True)
print(f'Number of rows after dropping duplicate replay tags: {combined_df.shape[0]}')
```

```python
combined_df.to_csv('../#2 Data Storage/Processed Data/preprocessed_battle_log_full_batch.csv', index=False)
```

```python
#delete the file which are output_path_csv
import os
from pathlib import Path

# delete the files which are in output_path_list (safe, non-failing)
deleted = []
failed = []
for p in output_path_list:
    fp = Path(p)
    try:
        if fp.exists():
            fp.unlink()
            deleted.append(str(fp))
        else:
            failed.append((str(fp), "not found"))
    except Exception as e:
        failed.append((str(fp), str(e)))

print(f"Deleted {len(deleted)} files")
for d in deleted:
    print(" -", d)
if failed:
    print(f"Failed to delete {len(failed)} files:")
    for fpath, err in failed:
        print(f" - {fpath}: {err}")
    
```
